```python
Q. Why LookupError class is used? Explain with an example KeyError and IndexError
```

```python
# This type of error is generally encountered while using maps.
# The LookupError raises two exceptions which are KeyError and IndexError.
# It serves as the base class for exceptions that result due to invalid keys or indexes used on a mapping or sequence.
```

```python
try :
    l = [1,2,3]
    logging.info(l[5])
except IndexError as e :
    logging.error("invalid index try again : {}".format(e))
```

```python
try :
    d = {"key1":1, "key2":2,"key3":3}
    logging.info(d["key4"])
except KeyError as e :
    logging.error("key not found :{}".format(e))
```
